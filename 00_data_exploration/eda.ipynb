```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/User/Documents/devanasokan_fyp/storage/cleanedlyrics.csv", encoding='utf-8-sig')
```

```python
# 1. Check if lyrics start with Track Name or Artist Name
def check_metadata_overlap(row):
    lyrics = str(row['lyrics']).lower()
    track = str(row['track_name']).lower()
    artist = str(row['artist_names']).lower().split('|')[0] # check first artist
    
    starts_with_track = lyrics.startswith(track)
    starts_with_artist = lyrics.startswith(artist)
    
    return starts_with_track or starts_with_artist
```

```python
df['metadata_overlap'] = df.apply(check_metadata_overlap, axis=1)
print(f"Number of songs starting with metadata: {df['metadata_overlap'].sum()}")
print(df[df['metadata_overlap']][['track_name', 'lyrics']].head())
```

```python
short_lyrics = df[df['lyrics'].str.len() < 50]
print(f"Number of very short lyrics: {len(short_lyrics)}")
print(short_lyrics[['track_name', 'lyrics']])
```

```python
# Check for leftovers of timestamps/brackets
leftovers = df['lyrics'].str.contains(r'\[|\]|\{|\}', na=False).sum()
print(f"Number of songs with remaining brackets: {leftovers}")
```

```python
# Check for instrumentals
instrumentals = df['lyrics'].str.contains(r'instrumental', case=False, na=False).sum()
print(f"Number of instrumental tags: {instrumentals}")
```

```python
# Check for "Lyrics by", "Produced by" etc again but as phrases
credits_count = df['lyrics'].str.contains(r'lyrics by|produced by|written by|writer\(s\)', case=False, na=False).sum()
print(f"Number of songs with credits: {credits_count}")
```
